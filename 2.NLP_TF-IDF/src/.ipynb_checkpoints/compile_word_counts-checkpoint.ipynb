```python
ts_word_count = df2.iloc[:,[0,-1]].dropna()
ts_word_count
```

```python
import math
def tfidf(word, pony, script):
    tf = script.loc[word, pony]
    num_said_word = script.loc[word, 'num_pony']
    idf = math.log10(6/num_said_word)
    return (tf*idf)
```

```python
ts_word_count['twilight sparkle'] = ts_word_count.index
ts_word_count['tfidf'] =ts_word_count['word'].apply(lambda x: tfidf(x, 'twilight sparkle', df2))
#ts_word_count_list = ts_word_count.word[1:5]
ts_word_count
```
